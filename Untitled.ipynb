```python
# # CatBoost vs. LightGBM: Full Benchmark (5 Datasets)
# **Hardware**: AMD Ryzen 7 5800H, NVIDIA RTX 3050 Ti  
# **Tasks**: Classification, Regression, Imbalanced Data...
# **Key Metrics**: Accuracy, RMSE, Precision-Recall....  

# ## the libraries needed for this are given below
# %%
# !pip install catboost lightgbm scikit-learn pandas numpy matplotlib seaborn imbalanced-learn xgboost

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, confusion_matrix, 
                            mean_squared_error, precision_recall_curve, 
                            average_precision_score)
from sklearn.datasets import fetch_california_housing
import time
import psutil
import cpuinfo
import subprocess
import lightgbm as lgb
# %% [markdown]
# ## 2. Load & Preprocess Datasets
# ### 2.1 Titanic (Classification)
# %%
titanic = pd.read_csv("train.csv")  
y_titanic = titanic['Survived']

gnment
X_titanic = titanic[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']].copy()

# Fill missing values
X_titanic['Age'] = X_titanic['Age'].fillna(X_titanic['Age'].median())
X_titanic['Embarked'] = X_titanic['Embarked'].fillna('S')

# Split
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(
    X_titanic, y_titanic, test_size=0.2, random_state=42)

# %% [markdown]
# ### 2.2 California Housing (Regression)
# %%
california = fetch_california_housing(as_frame=True)
X_california, y_california = california.data, california.target

# Split
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_california, y_california, test_size=0.2, random_state=42)

# %% [markdown]
# ### 2.3 Credit Card Fraud (Imbalanced Classification) *NEW*
# %%
fraud = pd.read_csv("creditcard.csv")  # Download from Kaggle
X_fraud = fraud.drop('Class', axis=1)
y_fraud = fraud['Class']

# Scale and split
scaler = StandardScaler()
X_fraud_scaled = scaler.fit_transform(X_fraud)
X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(
    X_fraud_scaled, y_fraud, test_size=0.2, random_state=42, stratify=y_fraud)

print("Fraud class distribution:\n", y_train_f.value_counts())

# ### 2.4 Wine Quality (Multiclass Classification)
# %%
from sklearn.datasets import fetch_openml
wine = fetch_openml(name='wine-quality-red', as_frame=True)
X_wine, y_wine = wine.data, wine.target.astype(int)

# Split
X_train_w, X_test_w, y_train_w, y_test_w = train_test_split(
    X_wine, y_wine, test_size=0.2, random_state=42)

print("Wine quality class distribution:\n", y_wine.value_counts())

# ### 2.5 Telco Customer Churn (Medium-Scale Binary)
# %%
telco = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Preprocess
X_telco = telco.drop(['customerID', 'Churn'], axis=1)
y_telco = telco['Churn'].map({'Yes': 1, 'No': 0})

# Handle categorical features
cat_cols = X_telco.select_dtypes(include=['object']).columns
for col in cat_cols:
    X_telco[col] = LabelEncoder().fit_transform(X_telco[col].astype(str))

# Handle missing values (if any) and convert TotalCharges to numeric
X_telco['TotalCharges'] = pd.to_numeric(X_telco['TotalCharges'], errors='coerce')
X_telco.fillna(0, inplace=True)

# Split
X_train_tel, X_test_tel, y_train_tel, y_test_tel = train_test_split(
    X_telco, y_telco, test_size=0.2, random_state=42, stratify=y_telco)

print("Churn class distribution:\n", y_train_tel.value_counts())
# %% [markdown]
# ## 3. GPU Configuration
# %%
# LightGBM GPU settings
lgb_gpu_params = {
    'device': 'gpu',
    'gpu_platform_id': 0,  # NVIDIA
    'gpu_device_id': 0,    # RTX 3050 Ti
    'verbose': -1
}

# CatBoost GPU settings (works best for regression)
cat_gpu_params = {
    'task_type': 'GPU',
    'devices': '0:1'
}

# Imbalance handling
fraud_params = {
    'scale_pos_weight': len(y_train_f[y_train_f==0]) / len(y_train_f[y_train_f==1]),  # CatBoost
    'is_unbalance': True  # LightGBM
}

# %% [markdown]
# ## 4. Model Training & Evaluation
# ### 4.1 Titanic (GPU-accelerated)
# %%
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier

# CatBoost
start = time.time()
cat_clf = CatBoostClassifier(
    iterations=200,
    cat_features=['Sex', 'Embarked'],
    **cat_gpu_params,
    verbose=0
)
cat_clf.fit(X_train_t, y_train_t)
cat_time_titanic = time.time() - start

# LightGBM (requires label encoding)

X_train_t_lgb = X_train_t.copy()
for col in ['Sex', 'Embarked']:
    le = LabelEncoder()
    X_train_t_lgb[col] = le.fit_transform(X_train_t_lgb[col])


X_test_t_lgb = X_test_t.copy()
for col in ['Sex', 'Embarked']:
    le = LabelEncoder()  # Reinitialize (or reuse the one from training)
    X_test_t_lgb[col] = le.fit_transform(X_test_t_lgb[col])  # Use same encoder



start = time.time()
lgb_clf = LGBMClassifier(
    n_estimators=200,
    **lgb_gpu_params
)
lgb_clf.fit(X_train_t_lgb, y_train_t)
lgb_time_titanic = time.time() - start

# Evaluate
y_pred_cat = cat_clf.predict(X_test_t)


print(f"CatBoost Accuracy: {accuracy_score(y_test_t, y_pred_cat):.4f}, Time: {cat_time_titanic:.2f}s")
print(f"LightGBM Accuracy: {accuracy_score(y_test_t, y_pred_lgb):.4f}, Time: {lgb_time_titanic:.2f}s")

# %% [markdown]
# ### 4.2 California Housing (GPU Regression)
# %%
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor

# CatBoost
start = time.time()
cat_reg = CatBoostRegressor(
    iterations=200,
    **cat_gpu_params,
    verbose=0
)
cat_reg.fit(X_train_c, y_train_c)
cat_time_california = time.time() - start

# LightGBM
start = time.time()
lgb_reg = LGBMRegressor(
    n_estimators=200,
    **lgb_gpu_params
)
lgb_reg.fit(X_train_c, y_train_c)
lgb_time_california = time.time() - start

# Evaluate
y_pred_cat_reg = cat_reg.predict(X_test_c)
y_pred_lgb_reg = lgb_reg.predict(X_test_c)

# With:
mse_cat = mean_squared_error(y_test_c, y_pred_cat_reg)
rmse_cat = np.sqrt(mse_cat)
print(f"CatBoost RMSE: {rmse_cat:.4f}, Time: {cat_time_california:.2f}s")

mse_lgb = mean_squared_error(y_test_c, y_pred_lgb_reg)
rmse_lgb = np.sqrt(mse_lgb)
print(f"LightGBM RMSE: {rmse_lgb:.4f}, Time: {lgb_time_california:.2f}s")

# %% [markdown]
# ### 4.3 Credit Card Fraud (Imbalanced GPU Training) *NEW*
# %%
# CatBoost
start = time.time()
cat_fraud = CatBoostClassifier(
    iterations=500,
    **cat_gpu_params,
    scale_pos_weight=fraud_params['scale_pos_weight'],
    verbose=0
)
cat_fraud.fit(X_train_f, y_train_f)
cat_time_fraud = time.time() - start

# LightGBM
start = time.time()
lgb_fraud = LGBMClassifier(
    n_estimators=500,
    **lgb_gpu_params,
    **{'is_unbalance': True}
)
lgb_fraud.fit(X_train_f, y_train_f)
lgb_time_fraud = time.time() - start

# Evaluate
y_pred_cat_f = cat_fraud.predict_proba(X_test_f)[:, 1]
y_pred_lgb_f = lgb_fraud.predict_proba(X_test_f)[:, 1]

print(f"CatBoost AP: {average_precision_score(y_test_f, y_pred_cat_f):.4f}, Time: {cat_time_fraud:.2f}s")
print(f"LightGBM AP: {average_precision_score(y_test_f, y_pred_lgb_f):.4f}, Time: {lgb_time_fraud:.2f}s")

# ### 4.4 Wine Quality (Multiclass)
# %%
from sklearn.metrics import balanced_accuracy_score

# CatBoost
start = time.time()
cat_wine = CatBoostClassifier(
    iterations=200,
    **cat_gpu_params,
    verbose=0
)
cat_wine.fit(X_train_w, y_train_w)
cat_time_wine = time.time() - start

# LightGBM
start = time.time()
lgb_wine = LGBMClassifier(
    n_estimators=200,
    **lgb_gpu_params
)
lgb_wine.fit(X_train_w, y_train_w)
lgb_time_wine = time.time() - start

# Evaluate
y_pred_cat_w = cat_wine.predict(X_test_w)
y_pred_lgb_w = lgb_wine.predict(X_test_w)

print(f"CatBoost Balanced Accuracy: {balanced_accuracy_score(y_test_w, y_pred_cat_w):.4f}, Time: {cat_time_wine:.2f}s")
print(f"LightGBM Balanced Accuracy: {balanced_accuracy_score(y_test_w, y_pred_lgb_w):.4f}, Time: {lgb_time_wine:.2f}s")
# ### 4.5 Telco Churn (Medium-Scale Binary)
# %%
# CatBoost
start = time.time()
cat_telco = CatBoostClassifier(
    iterations=200,
    **cat_gpu_params,
    verbose=0
)
cat_telco.fit(X_train_tel, y_train_tel)
cat_time_telco = time.time() - start

# LightGBM
start = time.time()
lgb_telco = LGBMClassifier(
    n_estimators=200,
    **lgb_gpu_params
)
lgb_telco.fit(X_train_tel, y_train_tel)
lgb_time_telco = time.time() - start

# Evaluate
y_pred_cat_tel = cat_telco.predict(X_test_tel)
y_pred_lgb_tel = lgb_telco.predict(X_test_tel)

print(f"CatBoost Accuracy: {accuracy_score(y_test_tel, y_pred_cat_tel):.4f}, Time: {cat_time_telco:.2f}s")
print(f"LightGBM Accuracy: {accuracy_score(y_test_tel, y_pred_lgb_tel):.4f}, Time: {lgb_time_telco:.2f}s")


#confusion matrix for titanic
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
sns.heatmap(confusion_matrix(y_test_t, y_pred_cat), 
            annot=True, fmt='d', ax=ax1, cmap='Blues')
ax1.set_title("CatBoost")

sns.heatmap(confusion_matrix(y_test_t, y_pred_lgb), 
            annot=True, fmt='d', ax=ax2, cmap='Reds')
ax2.set_title("LightGBM")
plt.suptitle("Titanic Survival Prediction")
plt.tight_layout()
plt.show()

# Precision-Recall Plot
precision_cat, recall_cat, _ = precision_recall_curve(y_test_f, y_pred_cat_f)
precision_lgb, recall_lgb, _ = precision_recall_curve(y_test_f, y_pred_lgb_f)

plt.figure(figsize=(8, 5))
plt.plot(recall_cat, precision_cat, label='CatBoost')
plt.plot(recall_lgb, precision_lgb, label='LightGBM')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Credit Card Fraud: Precision-Recall Curve')
plt.legend()
plt.show()

#california housing true vs predicted
plt.figure(figsize=(10, 5))
plt.scatter(y_test_c, y_pred_cat_reg, alpha=0.3, label='CatBoost')
plt.scatter(y_test_c, y_pred_lgb_reg, alpha=0.3, label='LightGBM')
plt.plot([y_test_c.min(), y_test_c.max()], 
         [y_test_c.min(), y_test_c.max()], 'k--')
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("California Housing Price Prediction")
plt.legend()
plt.show()
#wine quality
from sklearn.metrics import classification_report
print("CatBoost Classification Report:")
print(classification_report(y_test_w, y_pred_cat_w))

print("\nLightGBM Classification Report:")
print(classification_report(y_test_w, y_pred_lgb_w))
#tesco feature importance
# Telco Churn Feature Importance
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# CatBoost (correct method name)
cat_telco.get_feature_importance(prettified=True) \
         .set_index('Feature Id') \
         .plot(kind='barh', ax=ax1, title='CatBoost Feature Importance')

# LightGBM
lgb.plot_importance(lgb_telco, ax=ax2, title='LightGBM Feature Importance')

plt.tight_layout()
plt.show()
# %% [markdown]
# ## 5. Results Summary Table
# %%
results = pd.DataFrame({
    'Dataset': ['Titanic', 'California Housing', 'Credit Card Fraud', 'Wine Quality', 'Telco Churn'],
    'Task': ['Binary Classification', 'Regression', 'Imbalanced Binary', 'Multiclass', 'Medium-Scale Binary'],
    'CatBoost Time (s)': [cat_time_titanic, cat_time_california, cat_time_fraud, cat_time_wine, cat_time_telco],
    'LightGBM Time (s)': [lgb_time_titanic, lgb_time_california, lgb_time_fraud, lgb_time_wine, lgb_time_telco],
    'CatBoost Metric': [
        accuracy_score(y_test_t, y_pred_cat),
        np.sqrt(mean_squared_error(y_test_c, y_pred_cat_reg)),  # Manual RMSE
        average_precision_score(y_test_f, y_pred_cat_f),
        balanced_accuracy_score(y_test_w, y_pred_cat_w),
        accuracy_score(y_test_tel, y_pred_cat_tel)
    ],
    'LightGBM Metric': [
        accuracy_score(y_test_t, y_pred_lgb),
        np.sqrt(mean_squared_error(y_test_c, y_pred_lgb_reg)),  # Manual RMSE
        average_precision_score(y_test_f, y_pred_lgb_f),
        balanced_accuracy_score(y_test_w, y_pred_lgb_w),
        accuracy_score(y_test_tel, y_pred_lgb_tel)
    ],
    'Metric Name': ['Accuracy', 'RMSE', 'Average Precision', 'Balanced Accuracy', 'Accuracy']
})

print("### Extended Performance Summary:")
print(results.to_markdown(index=False))

cat_clf.save_model('catboost_titanic.cbm')
lgb_clf.booster_.save_model('lightgbm_titanic.txt')
cat_reg.save_model('catboost_california.cbm')
lgb_reg.booster_.save_model('lightgbm_california.txt')
```
